# Neccessary Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'Model': ['POP','ItemKNN', 'SLIM', 'SLIM_da'],
    'NDCG@10': [0.053, 0.162, 0.068, 0.061],
    'RecGap': [0.018, 0.014, 0.012, 0.005],
}

df = pd.DataFrame(data)
```

```python
# Set seaborn theme
sns.set_theme(style="whitegrid")

# Define shape mapping per model
marker_dict = {
    'SLIM': '>',
    'SLIM_da': '+',
    'ItemKNN': '^',
    'POP': 'D'
}

# Create the plot
plt.figure(figsize=(6, 4))
for model in df['Model']:
    subset = df[df['Model'] == model]
    plt.scatter(subset['NDCG@10'], subset['RecGap'], s=50, label=model, marker=marker_dict[model])

# Fit and plot trend line
sns.regplot(data=df, x='NDCG@10', y='RecGap', scatter=False, color='brown', line_kws={"linestyle": "--", "linewidth": 2}, ci=None)

# Formatting
plt.xlabel("NDCG@10")
plt.xlim(0, 0.3)
plt.ylim(0, 0.03)
plt.ylabel("RecGap")
plt.legend(title='Model', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title(r'$\it{LFM\text{-}1b\text{-}DemoBias_{sub}}$', fontsize=12)

plt.tight_layout()
plt.show()
```
